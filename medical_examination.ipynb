```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('medical_examination.csv')
df
```

```python
df
```

```python
df['height (m)'] = df['height'] / 100
df
```

```python
df['bmi'] = df['weight'] / (df['height (m)'] * df['height (m)'])
df
```

```python
df['overweight'] = df['bmi'].apply(lambda x: 1 if x > 25 else 0)
df
```

```python
df['cholesterol'] = df['cholesterol'].apply(lambda x: 1 if x > 1 else 0)
df
```

```python
df['gluc'] = df['gluc'].apply(lambda x: 1 if x > 1 else 0)
df
```

```python
df_long = pd.melt(df, id_vars=['cardio'], value_vars=['active', 'alco', 'cholesterol', 'gluc', 'overweight', 'smoke', ])
df_long
```

```python
sns.catplot(data=df_long, x='variable',
            hue='value', col='cardio', kind='count')
```

```python
df = df[(df['ap_lo'] <= df['ap_hi'])]
df
```

```python
df = df[(df['height'] >= df['height'].quantile(0.025))]
df
```

```python
(df['height'] < df['height'].quantile(0.975)).value_counts()
```

```python
df = df[(df['height'] < df['height'].quantile(0.975))]
df
```

```python
(df['weight'] > df['weight'].quantile(0.025)).value_counts()
```

```python
df = df[(df['weight'] > df['weight'].quantile(0.025))]
df
```

```python
(df['weight'] < df['weight'].quantile(0.975)).value_counts()
```

```python
df = df[(df['weight'] < df['weight'].quantile(0.975))]
df
```

```python
corr_matrix = df.corr()
```

```python
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
```

```python
f, ax = plt.subplots(figsize=(11, 9))
```

```python
sns.heatmap(corr_matrix, mask=mask, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```
